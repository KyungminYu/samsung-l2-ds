# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

# Solve

```python
df3 = df_origin.copy()
```

```python
df3['channel']
```

```python
df3['channel'].str.contains('Pixel|Oper').sum()
```

```python
df3 = df3.loc[~(df3['channel'].str.contains('Pixel|Oper')),]
df3
```

```python
df3['Netflix'] = df3['channel'].str.contains('Netflix') + 0
```

```python
df3['support_netflix'] = df3['channel'].str.contains('Netflix').astype(int)
df3 = df3.assign(support_prime_video = df3['channel'].str.contains('Prime Video').astype(int))
```

```python
df3['review_rate'] = df3['Reviews'] / df3['Ratings']
```

```python
df3['discount_rate'] = df3['current_price'] / df3['MRP']
```

```python
df3['is_high'] = df3['Picture_quality'].str.contains('4K|8K').astype(int)
```

```python
col_feature = ["review_rate", "MRP", "discount_rate", "support_netflix", "support_prime_video", "is_high"]
```

```python
df3 = df3.dropna()
df3
```

```python
model_rf = RandomForestRegressor(random_state=123)
```

```python
model_rf.fit(X=df3[col_feature], y=df3['Stars'])
```

```python
model_rf.feature_importances_
```

```python
col_feature
```

```python
pd.Series(index=col_feature, data=model_rf.feature_importances_).sort_values(ascending=False)
```
