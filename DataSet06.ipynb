```python
df = df_origin.copy()
```

```python
df.columns
```

```python
col_dummies = ['gender', 'relevant_experience', 'enrolled_university', 'education_level','major_discipline']
```

```python
df_dummies = pd.get_dummies(df[col_dummies])
df_dummies
```

```python
df_dummies.columns
```

```python
col_dummy_drop = ['gender_Other', 'relevant_experience_No relevant experience', 'enrolled_university_no_enrollment', 'education_level_Phd', 'major_discipline_STEM']
```

```python
df_dummies = df_dummies.drop(columns=col_dummy_drop)
df_dummies
df = df.drop(columns=col_dummies)
df
```

```python
job2 = pd.concat([df[['city_development_index', 'experience', 'last_new_job','training_hours', 'target', 'Xgrp']], df_dummies], axis = 1)
job2
```

```python
df = job2.copy()
```

```python
y = df.pop('target')
y
```

```python
X = df.drop(columns='Xgrp')
X
```

```python
modellg = LogisticRegression(random_state=123, C=100000, max_iter=1000, solver='liblinear')
```

```python
modellg.fit(X, y)
```

```python
modellg.coef_
```

```python
import numpy as np
np.exp(modellg.coef_).round(3).max().astype(str)[:-1]
```

# Q3

```python
df_train = job2.loc[job2['Xgrp'] == 'train', ]
df_train = df_train.drop(columns=['Xgrp'])
df_test = job2.loc[job2['Xgrp'] == 'test', ]
df_test = df_test.drop(columns=['Xgrp'])
df_train.shape, df_test.shape
```

```python
y_train = df_train.pop('target')
y_test = df_test.pop('target')
df_train.shape, df_test.shape
```

```python
model_knn = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski')
```

```python
model_knn.fit(df_train, y_train)
```

```python
y_pred = model_knn.predict(df_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
