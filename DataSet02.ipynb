# Q3

```python
 df_origin.columns
```

```python
col_regression = ['SeniorCitizen', 'Partner', 'Dependents', 'tenure', 'MonthlyCharges', 'TotalCharges', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingMovies']
target = 'Churn'
```

```python
df3 = df_origin.copy()
```

```python
df = pd.concat([df3[col_regression], df3[target]], axis=1)
df
```

```python
for col in col_regression:
    if (df[col].dtype == 'object'):
        df[col] = df[col].apply(lambda x : 1 if x == 'Yes' else (0 if x == 'No' else -1))
df[target] = df[target].apply(lambda x : 1 if x == 'Yes' else (0 if x == 'No' else -1))
df
```

# Q3 library

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
```

```python
min_max_scaler = MinMaxScaler()
```

```python
df
```

```python
scaler = min_max_scaler.fit(df)
scaler
```

```python
scaledArr = min_max_scaler.transform(df)
scaledArr
```

```python
df.columns
```

```python
df = pd.DataFrame(data=scaledArr, columns=df.columns, index=list(df.index.values))
df
```

```python
df_train, df_test = train_test_split(df,  test_size=0.3, random_state=123)
```

```python
df_test
```

```python
df_train
```

```python
model = LogisticRegression(random_state=123)
```

```python
x_test = df_test.loc[:, col_regression]
y_test = df_test.loc[:, target]
x_train = df_train.loc[:, col_regression]
y_train = df_train.loc[:, target]
```

```python
model.fit(x_train, y_train)
```

```python
res = model.score(x_test, y_test)
res
```

```python
y_pred = model.predict(x_test)
```

```python
f1_score(y_test, y_pred).round(2)
```
