# Q3

```python
df3 = df_origin.copy()
df3.shape
df3
```

```python
df3.columns
```

```python
df3['screen_size'].value_counts()
```

```python
df_screen_size_ohe = pd.get_dummies(df3['screen_size'])
df_exclude_screen_size = df3[['ROM', 'RAM', 'num_rear_camera', 'num_front_camera','battery_capacity', 'ratings', 'num_of_ratings', 'sales_price','discount_percent', 'sales']]
```

```python
df_exclude_screen_size
```

```python
df3 = pd.concat([df_screen_size_ohe, df_exclude_screen_size], axis=1)
df3
```

```python
df3.columns
```

```python
df3.shape
```

```python
col_X = ['Large', 'Medium', 'Small', 'Very Large', 'Very Small', 'ROM', 'RAM', 'num_rear_camera', 'num_front_camera', 'battery_capacity', 'ratings', 'num_of_ratings', 'sales_price', 'discount_percent']
col_Y = 'sales'
```

```python
X = df3[col_X]
y = df3[col_Y]
seed = 123
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size =0.8, random_state = seed)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
X_train_scaler = scaler.transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
neighbors = [3,5,7,9,11]
```

```python
targetNeighbor = -1
minScore = 1e9
for neighbor in neighbors:
    modelknn = KNeighborsRegressor(n_neighbors=neighbor)
    modelknn.fit(X_train_scaler, y_train)
    y_pred = modelknn.predict(X_test_scaler)
    score = mean_squared_error(y_test, y_pred)
    print(neighbor)
    print(score)    
    if (score < minScore):
        minScore = score
        targetNeighbor = neighbor

print("------------------")
print(targetNeighbor)
print(minScore)    
```
