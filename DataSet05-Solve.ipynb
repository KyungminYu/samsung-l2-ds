```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.tree import DecisionTreeRegressor
df_origin = pd.read_csv('../Datasets/card_cust.csv')
df_origin
```

```python
df_origin.isna().sum(axis = 0)
```

```python
df_origin['MINIMUM_PAYMENTS'] = df_origin['MINIMUM_PAYMENTS'].fillna(df_origin['MINIMUM_PAYMENTS'].mean())
```

```python
df = df_origin.copy()
```

```python
df[['CREDIT_LIMIT', 'BALANCE']].corr().iloc[0, 1]
```

```python
TENURES = df['TENURE'].unique()
```

```python
corrs=[]
for n in TENURES:
    _df = df.loc[df['TENURE'] == n, ['CREDIT_LIMIT', 'BALANCE']]
    _stat_corr = _df.corr().iloc[0, 1]
    corrs.append(_stat_corr)

round(max(corrs), 2)
```

# Q2

```python
df = df_origin.copy()
```

```python
s_cust_id = df.pop('CUST_ID')
s_cust_id
```

```python
df
```

```python
scaler = StandardScaler()
df_scaler = scaler.fit_transform(df)
```

```python
seed = 1234
```

```python
scores = []
for k in range(2, 6):
    model_km = KMeans(n_clusters=k, random_state=seed)
    model_km.fit(df_scaler)
    score = silhouette_score(df_scaler, model_km.labels_)
    
    scores = scores + [score]
```

```python
scores
```

```python
model_km = KMeans(n_clusters=2, random_state=seed)
model_km.fit(df_scaler)
```

```python
df['cluster'] = model_km.labels_
```

```python
df.groupby("cluster")['ONEOFF_PURCHASES'].mean().round(2)
```

```python
df_train = df[df['CUST_ID'] % 4 != 0]
df_test = df[df['CUST_ID'] % 4 == 0]
```

```python
y_train = df_train.pop('ONEOFF_PURCHASES')
y_test = df_test.pop('ONEOFF_PURCHASES')
```

## model

```python
model_dtr = DecisionTreeRegressor(random_state=seed)
```

```python
model_dtr.fit(df_train, y_train)
```

```python
y_pred = model_dtr.predict(df_test)
```
