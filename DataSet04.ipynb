# Q3

```python
df3 = df_origin.copy()
```

```python
df3
```

```python
df3 = df3[['user', 'prod', 'gender', 'age_group', 'job', 'city', 'marital', 'purchase']]
df3
```

```python
df3['gender'] = df3['gender'].replace('M', 1)
df3['gender'] = df3['gender'].replace('F', 0)
df3
```

```python
ages = df3['age_group'].unique()
ages
```

```python
idx = 0
for age in ages:
    df3['age_group'] = df3['age_group'].replace(age, idx)
    idx = idx + 1
    
df3
```

```python
prod_count = df3.groupby('user')['prod'].nunique()
prod_count
```

```python
purchace_sum = df3.groupby('user')['purchase'].sum()
purchace_sum
```

```python
df3 = df3.drop_duplicates(subset='user')
df3
```

```python
df3['job'] = df3['job'].astype('O')
```

```python
df3['job'].dtype
```

```python
df3 = df3[['user', 'gender', 'age_group', 'job', 'city', 'marital', 'purchase']]
df3
```

```python
df3 = pd.get_dummies(df3)
df3
```
